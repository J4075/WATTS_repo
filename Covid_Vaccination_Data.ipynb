```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/country_data/United%20States.csv')
df
```

```python
df = df.groupby(['vaccine'])
```

```python
df1 = df.loc[df['date'] >= "2021-07-18"]
df1
```

```python
sns.lineplot(data= df, x= "date",y= "total_vaccinations") 
plt.show()
```

```python
plt.scatter(x= df1["date"],y= df1["total_vaccinations"]) 
plt.title("Last month vaccination rate")
plt.show()
```
